### Traditional way to make **ChatBot**

```python
# user prompt questions
user_q1 = "what are the healthiest foods to eat every day ? "
user_q2 = "Make a recipe for those foods"
user_q3 = "What is the amount of Kcal ?"

# for system
system_propmt = "Answer as more detailed as possible."

#--------------- Q1 ------------------#
messages = [
    {"role":"system", "content":system_propmt}, # system prompt
    {"role":"user", "content":user_q1} # user prompt q1
]

response_q1 = openai.ChatCompletion.create(
            model = "gpt-3.5-turbo",
            messages = messages,
            temperature = .7,
            max_tokens = 1000
            )
response_q1 = response_q1["choices"][0]["message"]["content"]
print(user_q1,"\n",response_q1)

print("**"*60)
#--------------- Q2 ------------------# Adding Memort

messages = [
    {"role":"system", "content":system_propmt}, # system prompt
    {"role":"user", "content":user_q1}, # user prompt
    {"role":"assistant","content":response_q1}, # assistant prompt
    {"role":"user", "content":user_q2}  # user prompt q2
]

response_q2 = openai.ChatCompletion.create(
            model = "gpt-3.5-turbo",
            messages = messages,
            temperature = .7,
            max_tokens = 1000
            )

response_q2 = response_q2["choices"][0]["message"]["content"]
print(user_q2,"\n",response_q2)
print("**"*60)


#--------------- Q3 ------------------# Adding Memort

messages = [
    {"role":"system", "content":system_propmt}, # system prompt
    {"role":"user", "content":user_q1}, # user prompt q1
    {"role":"assistant","content":response_q1}, # assistant prompt
    {"role":"user", "content":user_q2}, # user prompt q2
    {"role":"assistant","content":response_q2}, # assistant prompt
    {"role":"user", "content":user_q3} # user prompt q3

]

response_q3 = openai.ChatCompletion.create(
            model = "gpt-3.5-turbo",
            messages = messages,
            temperature = .7,
            max_tokens = 1000
            )

response_q3 = response_q3["choices"][0]["message"]["content"]
print(user_q3,"\n",response_q3)
```

## Looping

```python
all_messages = list()
all_prompt = list()
model_response = list()

system_propmt = "Answer as concisely as possible."
all_messages.append({"role":"system", "content":system_propmt})

while True:
    user_prompt = input("You: ")
    if user_prompt.lower() in ["exit","quit"]:
        print("Thanks For Using My Chatbot ....")
        time.sleep(3)
        break
    
    # if user wronged and entered more space 
    elif re.sub(r'\s+', ' ', user_prompt)==" ":
        # `\s+` matches one or more whitespace characters (spaces, tabs, newlines)
        # `re.sub(r'\s+', '', user_prompt)` removes all whitespace
        # If the result is an empty string (""), it means the input contained only spaces
        continue  # Skip this iteration == entered new prompt
    
    else :
        ## append each user_prompt
        all_prompt.append(user_prompt)
        
        all_messages.append({"role":"user", "content":user_prompt})
    
        response = openai.ChatCompletion.create(
                                model = "gpt-3.5-turbo",
                                messages = all_messages,
                                temperature = .7,
                                max_tokens = 1000
                                )

        response = response["choices"][0]["message"]["content"]
        
        ## append each response
        model_response.append(response)

        print("ChatBot: ", response,"\n")

        ## add memory (answer of previous question )
        all_messages.append({"role":"assistant", "content":response})
```
